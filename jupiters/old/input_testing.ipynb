```python
from parrot import Parrot
import torch
import warnings
import pandas as pd
warnings.filterwarnings("ignore")
```

```python
def random_state(seed):
  torch.manual_seed(seed)
  if torch.cuda.is_available():
    torch.cuda.manual_seed_all(seed)

random_state(1234)
```

```python
parrot = Parrot(model_tag="prithivida/parrot_paraphraser_on_T5", use_gpu=False)
```

```python
df = pd.read_csv(dataset_path, sep=";", header=None)
display(df)
```

```python
samples1 = df[1:6]
samples2 = df[36:50]
samples3 = df[72:77]
samples4 = df[120:125]

print(samples2[1])
```

```python
def paraphrase(phrase):
    para_phrases = parrot.augment(input_phrase=phrase)
    if para_phrases == None:
      return None
    else:
      return para_phrases

def generate_sub_df(original, paraphrase, intent):
  new_row = {'Original': [original],
             'Paraphrase': [paraphrase],
             'Intent': [intent]}
  new_df = pd.DataFrame(new_row)
  return new_df

def generate_paraphrases(phrases):
  column_names = ['Original', 'Paraphrase', 'Intent']
  dataF = pd.DataFrame(columns = column_names)
  intent = phrases.iloc[0, 1]  
  print(f"Intent for this block is {intent}")
  sentences = phrases[0]
  for sentence in sentences:
    print("Working on new sentence...")
    paraphrases = paraphrase(sentence)  
    if paraphrases != None:      
      i = 0
      for p in paraphrases:
        if i < 3:
          dataF = dataF.append(generate_sub_df(sentence, p[0], intent))        
          i+=1
  return dataF  

def generate_paraphrased_dataset(samples):
  column_names = ['Original', 'Paraphrase', 'Intent']
  dataF = pd.DataFrame(columns = column_names)
  for sample in samples:
    dataF = dataF.append(generate_paraphrases(sample))
  return dataF        
```

```python
sample_list = [samples1, samples2, samples3, samples4]

new_df = generate_paraphrased_dataset(sample_list)
```

```python
intent1 = new_df.loc[new_df["Intent"] == "get_smells"]
intent2 = new_df.loc[new_df["Intent"] == "get_smells_date"]
intent3 = new_df.loc[new_df["Intent"] == "report"]
intent4 = new_df.loc[new_df["Intent"] == "info"]
```

```python

WEB_SERVICE_PATH = "https://sesacadocs.eu.ngrok.io/cadocsNLU/predict?message="
NUMBER_OF_SAMPLES = 3

import requests
import json
import random
```

Input testing steps (for each intent):
  - predicting the intent for both the original and the artificial sentence
  - ask the oracle if it is a PASS 
  

```python
def get_predictions(data):
  i = 0
  results = []

  for row in data.itertuples(index=False):
    if row[0] != row[1]: 
      i += 1
      if i > 3:
        pass
      else: 
        req1 = requests.get(WEB_SERVICE_PATH + "" + row[1])
        req2 = requests.get(WEB_SERVICE_PATH + "" + row[0])
        results.append({"Artificial_Text": req1.json().get("intent").get("text"), "Artificial_Intent": req1.json().get("intent").get("intent"), "Original_Text": req2.json().get("intent").get("text"), "Original_Intent": req2.json().get("intent").get("intent")})
  return results

def oracle(results):
  passes = 0

  for prediction in results:
    if (prediction.get("Artificial_Intent").get("name") == prediction.get("Original_Intent").get("name")) and (abs(float(prediction.get("Artificial_Intent").get("confidence")) - float(prediction.get("Original_Intent").get("confidence"))) < 0.2):
      passes += 1
  return passes
```

Intent 1: Get Community Smells

```python
res1 = get_predictions(intent1)
print(json.dumps(res1, indent=4, sort_keys=True))
passes1 = oracle(res1)
print("Intent 1 \nPassed "+str(passes1)+" on 3")
```

```python
res2 = get_predictions(intent2)
print(json.dumps(res2, indent=4, sort_keys=True))
passes2 = oracle(res2)
print("Intent 2 \nPassed "+str(passes2)+" on 3")
```
